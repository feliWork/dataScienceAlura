```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda1.head()
```

# 1. Análisis de facturación

```python
import matplotlib.pyplot as plt

facturacion = [
    tienda1["Precio"].sum(),
    tienda2["Precio"].sum(),
    tienda3["Precio"].sum(),
    tienda4["Precio"].sum()
]

tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]

plt.bar(tiendas, facturacion, color="skyblue")
plt.title("Facturación total por tienda")
plt.ylabel("Total en ventas (USD)")
plt.show()
```

La tienda 1 es la tienda que más factura en general, teniendo un total en ventas mucho mayor que cualqueier otra tienda.

# 2. Ventas por categoría

```python
def ventasCategoria(df, nombre):
    df["Categoría del Producto"].value_counts().head(5).plot(kind="bar", color="lightgreen")
    plt.title(f"Top 5 categorías - {nombre}")
    plt.ylabel("Cantidad vendida")
    plt.xticks(rotation=45)
    plt.show()

ventasCategoria(tienda1, "Tienda 1")
ventasCategoria(tienda2, "Tienda 2")
ventasCategoria(tienda3, "Tienda 3")
ventasCategoria(tienda4, "Tienda 4")
```

En general, todas las tiendas venden cosas muy parecidas, asi mismo, sus categorías más populares, son practicamente las mismas en todas las tiendas, siendo los muebles y electrónicos las categorías más vendidas en todas las tiendas.

# 3. Calificación promedio de la tienda

```python
calificaciones = [
    tienda1["Calificación"].mean(),
    tienda2["Calificación"].mean(),
    tienda3["Calificación"].mean(),
    tienda4["Calificación"].mean()
]

plt.bar(tiendas, calificaciones, color="orange")
plt.title("Calificación promedio por tienda")
plt.ylabel("Puntaje (1 a 5)")
plt.ylim(0, 5)
plt.show()
```

En general, el promedio de evaluación de clientes es muy igual en todas las tiendas, no obstante, la tienda 3, es la que mejor puntaje tiene entre todas las tiendas, puede que no sea la tienda que más vende, pero si es la que mejor calificación le dan los clientes.

# 4. Productos más y menos vendidos

```python
def productosVendidos(df, nombre):
    productos = df["Producto"].value_counts()
    print(f"{nombre} - Producto más vendido: {productos.idxmax()} ({productos.max()} ventas)")
    print(f"{nombre} - Producto menos vendido: {productos.idxmin()} ({productos.min()} venta)")

productosVendidos(tienda1, "Tienda 1")
productosVendidos(tienda2, "Tienda 2")
productosVendidos(tienda3, "Tienda 3")
productosVendidos(tienda4, "Tienda 4")
```

En la tienda 1, el microondas es el articulos más vendido, y los auriculares con microfono es el articulo menos vendido.

En la tienda 2, iniciando en programación es lo más vendido, mientras que los juegos de mesa, son muy poco vendidos.

En la tienda 3, el kit de bancas tiene las mayores ventas, y los bloques de construccion la menor cantidad de ventas

En la tienda 4, la cama box, tiene el mayor número de ventas y la guitarra eléctrica el menor.

# 5. Envío promedio por tienda

```python
envioPromedio = [
    tienda1["Costo de envío"].mean(),
    tienda2["Costo de envío"].mean(),
    tienda3["Costo de envío"].mean(),
    tienda4["Costo de envío"].mean()
]

plt.bar(tiendas, envioPromedio, color="purple")
plt.title("Costo de envío promedio por tienda")
plt.ylabel("Costo (USD)")
plt.show()
```

En general, al igual que en su facturación total, la tienda 1 tiene el mayor promedio de costo de envío, por lo que puede que al estar vendiendo más, este valor corresponda a ser el más alto.
